```python
# for TEMPO NO2; to run for HCHO change tempo_l2_no2 to tempo_l2_hcho

for d in range(20,21):
    an = driver.analysis()

    an.control = 'control_camx_tempo.yaml'

    an.read_control()
    an.control_dict["analysis"]["start_time"] = f"2024-07-{d:02d}"
    an.control_dict["analysis"]["end_time"] = f"2024-07-{d:02d} 23:59:00"

    modfile = an.control_dict["model"]["CAMX"]["files"].replace(
        "2024-07-20", f"2024-07-{d:02d}"
    )
    modfile = modfile.replace("2024-07-20", f"2024-07-{d:02d}")
    print (modfile)

    an.control_dict["model"]["CAMX"]["files"] = modfile
    satfile = an.control_dict["obs"]["tempo_l2_hcho"]["filename"].replace(
        "20240720", f"202407{d:02d}"
    )
    print (satfile)
    an.save["paired"]["prefix"] = f"2024-07-{d:02d}_filtered"
    an.start_time = pd.Timestamp(an.control_dict["analysis"]["start_time"])
    an.end_time = pd.Timestamp(an.control_dict["analysis"]["end_time"])
    an.control_dict["obs"]["tempo_l2_hcho"]["filename"] = satfile
    an.open_models()
    an.open_obs()
#this line is only needed to adjust the CAMx date from 2022 to 2024 since no TEMPO data are available for 2022
    an.models["CAMX"].obj["time"] = an.models["CAMX"].obj["time"]  + np.timedelta64(731, "D")
    an.pair_data()
    an.save["paired"]["prefix"] = f"2024-07-{d:02d}"   #set prefix for simulation from yaml file here
    an.save_analysis()
    gc.collect()
```
